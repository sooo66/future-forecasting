```python
import json
# read_jsonl
def read_jsonl(file_path):
    data = []
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            data.append(json.loads(line))
    return data
data = read_jsonl("/Users/ymx66/Workspace/future-forecasting/data/url_pool/url_pool.jsonl")
# 统计每个domain的数量
from collections import Counter
domains = [item["domain"] for item in data]
domain_counts = Counter(domains)
print(domain_counts)
total_count = sum(domain_counts.values())
print("Total count:", total_count)

# 除去 bloomberg, forbes, nytimes, theepochtimes 算总数
excluded_domains = {"bloomberg.com", "forbes.com", "nytimes.com", "theepochtimes.com"}
filtered_count = sum(count for domain, count in domain_counts.items() if domain not in excluded_domains)
print("Filtered count (excluding bloomberg, forbes, nytimes, theepochtimes):", filtered_count)
```
